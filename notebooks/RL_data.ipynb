```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("rl_training_log.csv")
df.head()
```

```python
plt.plot(df['batch'][::10], df['avg_reward'][::10])
```

```python
fig, ax1 = plt.subplots()

# 3. Instantiate a second axes that shares the same x-axis (right y-axis)
ax2 = ax1.twinx() 

# 4. Plot the first dataset on the left axis and label it
color1 = 'tab:red'
ax1.set_xlabel('Batches')
ax1.set_ylabel('Average Reward', color=color1)
ax1.plot(df['batch'][::10], df['avg_reward'][::10], color=color1)
ax1.tick_params(axis='y', labelcolor=color1)

# 5. Plot the second dataset on the right axis and label it
color2 = 'tab:blue'
ax2.set_ylabel('KL Divergence', color=color2)
ax2.plot(df['batch'][::10], df['kl_div'][::10], color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

# 6. Optional: Add a title and ensure a tight layout
plt.title('Reinforcement Learning Log')
fig.tight_layout() # Ensures labels don't get clipped

# 7. Display the plot
plt.show()
```
